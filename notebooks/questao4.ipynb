# Questão 4 — Análise da energia da imagem no domínio da DCT

Este notebook apresenta a implementação em Python utilizada para resolver a Questão 4 da Aula Prática 4 de Processamento de Sinais I.

## Objetivo

Aplicar a **DCT bidimensional** à imagem `sosias.jpg` e analisar a distribuição de energia no domínio da transformada.

O objetivo é verificar como a energia da imagem se concentra nas componentes de baixa e alta frequência.

## Bibliotecas utilizadas

Serão utilizadas as bibliotecas:

- `NumPy` para cálculos numéricos;
- `Matplotlib` para geração de gráficos;
- `Pillow` para leitura da imagem;
- `SciPy` para aplicação da DCT 2-D.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.fft import dctn
```

## Leitura da imagem

A imagem `sosias.jpg` será carregada e convertida para escala de cinza, de modo que a análise seja feita sobre uma matriz bidimensional de intensidades.

```python
img = Image.open("../dados/sosias.jpg").convert("L")
A = np.array(img, dtype=float)

print("Dimensões da imagem:", A.shape)
```

## Visualização da imagem original

A seguir, é exibida a imagem em escala de cinza que será utilizada na análise.

```python
plt.figure(figsize=(8, 5))
plt.imshow(A, cmap="gray")
plt.axis("off")
plt.title("Imagem em escala de cinza")
plt.tight_layout()
plt.show()
```

## Aplicação da DCT 2-D

A DCT bidimensional será aplicada à matriz da imagem para obter sua representação no domínio da frequência espacial.

A energia total da imagem no domínio da transformada é dada por

$$
E = \sum_u \sum_v |C(u,v)|^2,
$$

em que \(C(u,v)\) representa os coeficientes da DCT 2-D.

```python
C = dctn(A, type=2, norm="ortho")
E = np.abs(C) ** 2
E_total = E.sum()

print("Energia total da imagem no domínio da DCT:", E_total)
```

## Visualização da magnitude da DCT

Como os valores dos coeficientes da DCT podem variar bastante, é conveniente visualizar a magnitude da transformada em escala logarítmica, isto é,

$$
\log(1 + |C(u,v)|).
$$

Essa visualização ajuda a identificar em quais regiões da transformada a energia está mais concentrada.

```python
logC = np.log1p(np.abs(C))

plt.figure(figsize=(8, 5))
plt.imshow(logC, cmap="gray")
plt.axis("off")
plt.title("Magnitude logarítmica da DCT 2-D")
plt.tight_layout()
plt.show()
```

## Energia acumulada nas baixas frequências

Como as componentes de baixa frequência estão concentradas no canto superior esquerdo da matriz da DCT, será calculada a energia acumulada em blocos

$$
K \times K
$$

nessa região.

A ideia é verificar quanto da energia total da imagem é preservado à medida que o tamanho do bloco aumenta.

```python
Ks = np.array([4, 8, 16, 32, 64, 128, 256, min(A.shape)])
cum_energy = []

for K in Ks:
    energy_block = E[:K, :K].sum()
    cum_energy.append(100 * energy_block / E_total)

cum_energy = np.array(cum_energy)

for K, val in zip(Ks, cum_energy):
    print(f"Bloco {K}x{K}: {val:.4f}% da energia total")
```

## Gráfico da energia acumulada

O gráfico abaixo mostra a porcentagem da energia total concentrada no bloco \(K \times K\) do canto superior esquerdo da DCT.

Se a energia crescer rapidamente para valores pequenos de \(K\), isso indica forte concentração nas baixas frequências.

```python
plt.figure(figsize=(8, 5))
plt.plot(Ks, cum_energy, marker='o')
plt.xlabel('Tamanho do bloco no canto superior esquerdo (K x K)')
plt.ylabel('Energia acumulada (%)')
plt.title('Energia acumulada da DCT em baixas frequências')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Interpretação dos resultados

A análise da energia acumulada permite verificar se a imagem apresenta forte concentração de energia nas componentes de baixa frequência.

Esse comportamento é esperado em imagens naturais, pois grande parte da informação visual costuma estar associada à estrutura global da cena, enquanto os detalhes finos e as transições bruscas aparecem nas componentes de alta frequência.

## Conclusão

Os resultados mostram que a maior parte da energia da imagem está concentrada nos coeficientes localizados próximos ao canto superior esquerdo da DCT, isto é, nas componentes de baixa frequência.

Isso significa que a imagem pode ser representada de forma eficiente por um conjunto relativamente pequeno de coeficientes da transformada, preservando boa parte de sua informação visual global.

Esse comportamento explica por que a DCT é amplamente utilizada em técnicas de compressão de imagens: ela apresenta alta capacidade de compactação energética e permite representar a estrutura principal da imagem com poucos coeficientes.
