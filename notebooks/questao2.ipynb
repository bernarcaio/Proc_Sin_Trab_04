# Questão 2 — Resolução espectral e efeito do zero-padding

Este notebook apresenta a implementação em Python utilizada para resolver a Questão 2 da Aula Prática 4 de Processamento de Sinais I.

## Objetivo

Analisar o impacto do tamanho da DFT na resolução espectral do sinal

$$
x(t) = \sin(2\pi t) + \sin(2{,}02\pi t),
$$

amostrado com frequência

$$
f_s = 10\ \text{Hz}.
$$

O estudo compara o espectro obtido com diferentes quantidades de amostras reais e diferentes tamanhos de DFT, incluindo casos com zero-padding.

## Bibliotecas utilizadas

Serão utilizadas as bibliotecas:

- `NumPy` para cálculos numéricos;
- `Matplotlib` para geração dos gráficos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definição do sinal

O sinal contínuo é dado por

$$
x(t) = \sin(2\pi t) + \sin(2{,}02\pi t).
$$

Comparando com a forma geral

$$
\sin(2\pi f t),
$$

observa-se que o sinal é composto por duas senoides com frequências

$$
f_1 = 1\ \text{Hz}
\qquad \text{e} \qquad
f_2 = 1{,}01\ \text{Hz}.
$$

A diferença entre elas é

$$
\Delta f = 1{,}01 - 1 = 0{,}01\ \text{Hz}.
$$

## Resolução em frequência

A resolução em frequência associada a uma DFT de \(N\) pontos é dada por

$$
\Delta f_{\text{DFT}} = \frac{f_s}{N}.
$$

Entretanto, é importante destacar que o uso de zero-padding apenas torna a amostragem do espectro mais densa, mas não aumenta a resolução espectral real do sinal. A melhora real da capacidade de distinguir componentes próximas depende do aumento do número de amostras reais, isto é, do tempo de observação do sinal.

```python
fs = 10.0

def generate_signal(num_samples):
    n = np.arange(num_samples)
    t = n / fs
    x = np.sin(2*np.pi*1.0*t) + np.sin(2*np.pi*1.01*t)
    return n, t, x
```

## Função para cálculo do espectro

A célula abaixo define uma função para calcular a DFT e retornar apenas o semieixo positivo de frequências, facilitando a visualização dos resultados em Hz.

```python
def spectrum(x, Nfft, fs):
    X = np.fft.fft(x, n=Nfft)
    f = np.fft.fftfreq(Nfft, d=1/fs)
    mask = f >= 0
    return f[mask], np.abs(X[mask])
```

## Caso 1 — 64 amostras reais e DFT de 64 pontos

Neste caso, o tempo de observação é

$$
T = \frac{64}{10} = 6{,}4\ \text{s}
$$

e o espaçamento entre bins da DFT é

$$
\Delta f_{\text{DFT}} = \frac{10}{64} = 0{,}15625\ \text{Hz}.
$$

Como a separação entre as senoides é apenas \(0{,}01\ \text{Hz}\), espera-se que as duas componentes não possam ser distinguidas claramente.

```python
_, _, x64 = generate_signal(64)
f64, mag64 = spectrum(x64, 64, fs)

plt.figure(figsize=(9, 5))
plt.plot(f64, mag64)
plt.xlim(0, 2.0)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude da DFT')
plt.title('64 amostras reais, DFT de 64 pontos')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Caso 2 — 64 amostras reais e DFT de 128 pontos com zero-padding

Agora são mantidas as mesmas 64 amostras reais, mas a DFT é calculada com 128 pontos, por meio da adição de 64 zeros.

O novo espaçamento entre bins é

$$
\Delta f_{\text{DFT}} = \frac{10}{128} = 0{,}078125\ \text{Hz}.
$$

Apesar de a malha de frequências ficar mais densa, isso não significa aumento da resolução real, pois o tempo de observação do sinal continua o mesmo.

```python
f64zp, mag64zp = spectrum(x64, 128, fs)

plt.figure(figsize=(9, 5))
plt.plot(f64zp, mag64zp)
plt.xlim(0, 2.0)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude da DFT')
plt.title('64 amostras reais, DFT de 128 pontos com zero-padding')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Caso 3 — 128 amostras reais e DFT de 128 pontos

Neste caso, o tempo de observação passa a ser

$$
T = \frac{128}{10} = 12{,}8\ \text{s},
$$

e a resolução em frequência é

$$
\Delta f_{\text{DFT}} = \frac{10}{128} = 0{,}078125\ \text{Hz}.
$$

Embora o valor numérico do espaçamento seja igual ao do caso anterior com zero-padding, agora há mais informação real no sinal, pois ele foi observado durante mais tempo.

```python
_, _, x128 = generate_signal(128)
f128, mag128 = spectrum(x128, 128, fs)

plt.figure(figsize=(9, 5))
plt.plot(f128, mag128)
plt.xlim(0, 2.0)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude da DFT')
plt.title('128 amostras reais, DFT de 128 pontos')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Casos com zero-padding adicional

A seguir, são analisados dois casos adicionais:

- 128 amostras reais com DFT de 256 pontos;
- 128 amostras reais com DFT de 512 pontos.

Os respectivos espaçamentos entre bins são

$$
\frac{10}{256} = 0{,}0390625\ \text{Hz}
$$

e

$$
\frac{10}{512} = 0{,}01953125\ \text{Hz}.
$$

Mais uma vez, o zero-padding melhora a visualização do espectro, mas não acrescenta nova informação ao sinal.

```python
f256, mag256 = spectrum(x128, 256, fs)
f512, mag512 = spectrum(x128, 512, fs)

plt.figure(figsize=(9, 5))
plt.plot(f256, mag256, label='DFT de 256 pontos')
plt.plot(f512, mag512, label='DFT de 512 pontos')
plt.xlim(0, 2.0)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude da DFT')
plt.title('128 amostras reais com zero-padding')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Comparação geral dos cinco casos

Para facilitar a análise, a célula seguinte gera todos os casos em sequência:

1. 64 amostras reais, DFT de 64 pontos;
2. 64 amostras reais, DFT de 128 pontos com zero-padding;
3. 128 amostras reais, DFT de 128 pontos;
4. 128 amostras reais, DFT de 256 pontos com zero-padding;
5. 128 amostras reais, DFT de 512 pontos com zero-padding.

```python
cases = [
    ("64 amostras reais, DFT de 64 pontos", x64, 64),
    ("64 amostras reais, DFT de 128 pontos com zero-padding", x64, 128),
    ("128 amostras reais, DFT de 128 pontos", x128, 128),
    ("128 amostras reais, DFT de 256 pontos com zero-padding", x128, 256),
    ("128 amostras reais, DFT de 512 pontos com zero-padding", x128, 512),
]

for title, signal, Nfft in cases:
    f, mag = spectrum(signal, Nfft, fs)

    plt.figure(figsize=(9, 5))
    plt.plot(f, mag)
    plt.xlim(0, 2.0)
    plt.xlabel('Frequência (Hz)')
    plt.ylabel('Magnitude da DFT')
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## Conclusão

Os resultados mostram que o aumento do tamanho da DFT por zero-padding torna o espectro visualmente mais detalhado, pois aumenta a densidade de amostragem no eixo da frequência. No entanto, essa técnica não melhora a resolução espectral real do sinal.

A melhora real da capacidade de distinguir componentes muito próximas depende do aumento do número de amostras reais, isto é, do aumento do tempo de observação do sinal.

Neste problema, como as senoides possuem frequências de \(1\ \text{Hz}\) e \(1{,}01\ \text{Hz}\), a separação entre elas é muito pequena, de modo que mesmo com 128 amostras reais ainda não é possível distingui-las de forma nítida. Assim, conclui-se que o zero-padding é útil para interpolação espectral e melhor visualização, mas não substitui o aumento real da duração do sinal quando o objetivo é melhorar a resolução em frequência.
