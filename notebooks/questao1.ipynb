# Questão 1 — Comparação entre DTFT e DFT

Este notebook apresenta a implementação em Python utilizada para resolver a Questão 1 da Aula Prática 4 de Processamento de Sinais I.

## Objetivo

Comparar a DTFT do sinal

$$
x[n] = \delta[n] - \delta[n-1] + \delta[n-2] - \delta[n-3]
$$

com a DFT para diferentes valores de \(N\), analisando como o aumento do número de pontos melhora a representação espectral.

## Bibliotecas utilizadas

Serão utilizadas as bibliotecas:

- `NumPy` para cálculos numéricos;
- `Matplotlib` para geração dos gráficos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definição do sinal

O sinal dado na questão é

$$
x[n] = \delta[n] - \delta[n-1] + \delta[n-2] - \delta[n-3]
$$

Como ele é não nulo apenas em \(n = 0,1,2,3\), pode ser representado como

$$
x[n] = \{1,-1,1,-1\}.
$$

```python
x = np.array([1, -1, 1, -1], dtype=complex)
print("Sinal x[n] =", x)
```

## Cálculo da DTFT

A DTFT será calculada em uma malha densa de frequências angulares no intervalo

$$
0 \leq \omega < 2\pi
$$

para que sua representação seja praticamente contínua.

A definição utilizada é

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n]e^{-j\omega n}.
$$

```python
w = np.linspace(0, 2*np.pi, 4000, endpoint=False)
X_dtft = np.zeros_like(w, dtype=complex)

for n, xn in enumerate(x):
    X_dtft += xn * np.exp(-1j * w * n)
```

## Gráfico da DTFT

Abaixo é mostrado o gráfico da magnitude da DTFT do sinal, isto é,

$$
|X(e^{j\omega})|.
$$

```python
plt.figure(figsize=(9, 5))
plt.plot(w, np.abs(X_dtft))
plt.xlabel('Frequência angular ω (rad/amostra)')
plt.ylabel('Magnitude')
plt.title('Magnitude da DTFT do sinal x[n]')
plt.xlim(0, 2*np.pi)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', 'π/2', 'π', '3π/2', '2π'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Cálculo e comparação com a DFT

Agora será calculada a DFT para os valores

$$
N = 4,\;16,\;64,\;1024
$$

A ideia é comparar os pontos discretos da DFT com a curva contínua da DTFT.

A DFT é avaliada nas frequências

$$
\omega_k = \frac{2\pi k}{N}, \qquad k = 0,1,\dots,N-1.
$$

```python
Ns = [4, 16, 64, 1024]

for N in Ns:
    k = np.arange(N)
    wk = 2 * np.pi * k / N
    X_dft = np.fft.fft(x, n=N)

    plt.figure(figsize=(9, 5))
    plt.plot(w, np.abs(X_dtft), label='|DTFT|')
    plt.stem(wk, np.abs(X_dft), linefmt='C1-', markerfmt='C1o',
             basefmt=' ', label=f'|DFT|, N={N}')

    plt.xlabel('Frequência angular ω (rad/amostra)')
    plt.ylabel('Magnitude')
    plt.title(f'Comparação entre DTFT e DFT para N = {N}')
    plt.xlim(0, 2*np.pi)
    plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
               ['0', 'π/2', 'π', '3π/2', '2π'])
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## Verificação numérica para \(N = 4\)

No caso \(N = 4\), espera-se obter

$$
X[k] = \{0,0,4,0\}.
$$

A célula abaixo mostra esse resultado numericamente.

```python
X4 = np.fft.fft(x, n=4)
print("DFT para N=4:")
print(X4)
print("Magnitude:")
print(np.abs(X4))
```

## Conclusão

Os gráficos mostram que a DTFT representa o espectro contínuo do sinal, enquanto a DFT corresponde à amostragem desse espectro em frequências discretas.

Para \(N=4\), a representação é bastante limitada. À medida que \(N\) aumenta para \(16\), \(64\) e \(1024\), os pontos da DFT tornam-se mais densos e passam a se aproximar melhor da curva da DTFT.

Assim, conclui-se que o aumento de \(N\) melhora a representação espectral do sinal no domínio da frequência, embora não altere o conteúdo espectral do sinal original.
